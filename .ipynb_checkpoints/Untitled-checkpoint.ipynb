```python
import pandas as pd
import numpy as np
import os

construction_path = os.getcwd() + '\data\CONSTRUCTION_LOCATIONS.npy'
construction_dict = np.load(construction_path,allow_pickle=True).item()
construction_df = pd.DataFrame(construction_dict)
construction_df = construction_df.transpose().rename(columns={0:'Address', 1:'Latitude', 2:'Longitude'}).reset_index().drop(columns=['index'])
construction_df
```

```python
import folium
import folium.plugins # The Folium Javascript Map Library

NY_COORDINATES = (40.785091, -73.968285)

sf_map = folium.Map(NY_COORDINATES, zoom_start=13)
locs = construction_df[['Latitude','Longitude']].dropna() #get the latitude and longitude, drop all null values
heatmap = folium.plugins.HeatMap(locs, radius=10)

sf_map.add_child(heatmap)
```
